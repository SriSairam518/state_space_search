```python
class State:
    def __init__(self,state,umbrella_pos,required_times,time_taken):
        self.state = state
        self.umbrella_pos = umbrella_pos
        self.required_times = required_times
        self.time_taken = time_taken
    
    def goalTest(self):
        # ['W','W','W','W']
        goal = [1,1,1,1]
        return self.state == goal

    def moveGen(self):
        children = []
        for i in range(len(self.state)):
            for j in range(len(self.state)):
                #position of the umbrella to be moved
                # new_position = 'W' if self.umbrella_pos == 'E' else 'E'
                new_position = 1 if self.umbrella_pos == 0 else 0

                if i == j and self.umbrella_pos == self.state[i]:
                    new_state = self.state.copy()
                    new_state[i] = new_position
                    children.append(State(new_state, new_position, self.required_times, self.time_taken+self.required_times[i]))
                elif self.umbrella_pos == self.state[i] == self.state[j]:
                    new_state = self.state.copy()
                    new_state[i] = new_state[j] = new_position
                    children.append(State(new_state, new_position, self.required_times, self.time_taken+max(self.required_times[i],self.required_times[j])))
                
        return children
    
    def __eq__(self,other):
        return isinstance(other, State) and self.state == other.state and self.umbrella_pos == other.umbrella_pos
    
    def __str__(self):
        return f'--> [ {self.state[0]}, {self.state[1]}, {self.state[2]}, {self.state[3]} ] , umbrella position : {self.umbrella_pos} , time taken : {self.time_taken}'

    def __hash__(self):
        return hash(tuple(self.state))

# state = State([0,0,0,0],0,[5,10,20,25],0)
# print("State : " ,state)

# print("goal test : ",state.goalTest())

# children = state.moveGen()
# for i in children:
#     print(i)
```

```python
def filterStates(children,open,closed):
    open_states = [o for o,parent in open]
    closed_states = [c for c,parent in closed]
    new_states = [state for state in children if state not in open_states and state not in closed_states and state.time_taken<=60]
    return new_states

def constructPath(pair,closed):
    path = []
    parent_map = {}
    
    for node,parent in closed:
        parent_map[node] = parent
    
    node,parent = pair
    path.append(node)
    while parent:
        path.append(parent)
        parent = parent_map[parent]
    
    path.reverse()
    return path

def printPath(path):
    for i in range(len(path)):
        print(i,"*",path[i])

def BFS(start):
    open = [(start,None)]
    closed = []
    while open:
        pair = open.pop(0)
        n,parent = pair
        if n.goalTest():
            path = constructPath(pair,closed)
            printPath(path)
            return
        closed.append(pair)
        children = n.moveGen()
        updated_children = filterStates(children,open,closed)
        new_states = [(state,n) for state in updated_children]
        open = open + new_states
    return []

state = State([0,0,0,0],0,[5,10,20,25],0)
BFS(state)
```
