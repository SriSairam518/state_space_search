```python
class State:
  def __init__(self,state):
    self.state=state

  def goalTest(self):
    # state = ['w','w','w','-','e','e','e']
    # state2 = ['w','w','w','-','e','e','e']
    # print(state==state2) True
    # goal = [2,2,2,0,1,1,1]
    goal = ['W','W','W','_','E','E','E']
    return self.state==goal

  def moveGen(self):
    children = []
    for i in range(len(self.state)):
      if self.state[i]=='E':
        if i+1<len(self.state) and self.state[i+1]=='_':
          new_state = self.state.copy()
          new_state[i],new_state[i+1] = '_','E'
          children.append(State(new_state))
        elif i+2<len(self.state) and self.state[i+2]=='_':
          new_state=self.state.copy()
          new_state[i],new_state[i+2] = '_','E'
          children.append(State(new_state))
      elif self.state[i]=='W':
        if i-1>=0 and self.state[i-1]=='_':
          new_state = self.state.copy()
          new_state[i],new_state[i-1] = '_','W'
          children.append(State(new_state))
        elif i-2>=0 and self.state[i-2]=='_':
          new_state = self.state.copy()
          new_state[i],new_state[i-2] = '_','W'
          children.append(State(new_state))
    return children

  def __eq__(self,other):
    return self.state==other.state

  def __str__(self):
    return f'--> {self.state[0]}, {self.state[1]}, {self.state[2]}, {self.state[3]}, {self.state[4]}, {self.state[5]}, {self.state[6]} '

  def __hash__(self):
    return hash(tuple(self.state))

state1 = State([1,1,1,0,2,2,2])
# print(state1)

state2 = State([1,1,1,0,2,2,2])
# print(state2)

# print("is equal : ", state1==state2)

children = state1.moveGen()
for i in children:
  print(i)
```

```python
def removeSeen(children,open,closed):
  open_states = [o for o,parent in open]
  closed_states = [c for c,parent in closed]
  new_states = [new_state for new_state in children if new_state not in open_states and new_state not in closed_states]
  return new_states

def constructPath(pair,closed):
  path = []
  parent_map = {}

  for node,parent in closed:
    parent_map[node] = parent

  node,parent = pair
  path.append(node)
  while parent:
    path.append(parent)
    parent=parent_map[parent]
  path.reverse()
  return path

def printPath(path):
  for i in range(len(path)):
    print(i,"*",path[i])

def bfs(start):
  open = [(start,None)]
  closed = []
  while open:
    pair = open.pop(0)
    n,parent = pair

    # print(n)
    if n.goalTest():
      path = constructPath(pair,closed)
      printPath(path)
      return
    else:
      closed.append(pair)
      children = n.moveGen()
      updated_children = removeSeen(children,open,closed)
      new_children = [(state,n) for state in updated_children]
      open = open + new_children
  return []

def dfs(start):
  open = [(start,None)]
  closed = []
  while open:
    pair = open.pop(0)
    n,parent = pair
    if n.goalTest():
      path = constructPath(pair,closed)
      printPath(path)
      return
    closed.append(pair)
    children = n.moveGen()
    updated_children = removeSeen(children,open,closed)
    new_children = [(state,n) for state in updated_children]
    open = new_children + open
  return []


state1 = State(['E','E','E','_','W','W','W'])
print("USING BFS : ")
bfs(state1)

print()

print("USING DFS : ")
dfs(state1)
```
